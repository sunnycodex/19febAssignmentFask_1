```python
#2.
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#3. App routing in Flask is the process of mapping URLs to specific functions that handle the logic for those URLs. For example, if you want to display a welcome message when someone visits your website’s homepage, you can use the @app.route('/') decorator to associate the URL / with a function that returns the message.
#We use app routes in Flask to make our web applications more user-friendly and easy to navigate. App routes allow us to define meaningful URLs that correspond to different functions or views in our code.
```

```python
#4.from flask import Flask

app = Flask(__name__)

@app.route("/")
def sunny():
    return "<h1>Company Name: ABC Corporation</h1> <h1>Location: India</h1> <h1>Contact Detail: 999-999-9999</h1>"

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#5. url_for() used in Flask for url bulding.
from flask import Flask, url_for
app = Flask(__name__)

@app.route('/')
def home():
    return "Welcome to the pythongeeks page!"

@app.route('/about')
def about():
    return "This is the about page."

@app.route('/profile/<username>')
def profile(username):
    return f"Welcome to the profile page of {username}."

if __name__ == '__main__':
    app.run(debug=True)
```
